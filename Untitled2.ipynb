```python
import sys, time
from PyQt5.QtGui import QPainter, QColor, QBrush, QPixmap
from PyQt5.QtWidgets import (QLineEdit, QSlider, QPushButton, QHBoxLayout,
                             QVBoxLayout, QApplication, QWidget, QLabel)
from PyQt5.QtCore import Qt


class MainScreen(QWidget):
    
    # Creating a Method to initialize the main screen
    def __init__(self):
        super().__init__()
        
        self.width = 2600
        self.height = 1600
        self.init_ui()
        
    
    def init_ui(self):
        
        """
        w = QtWidgets.QWidget()
    
        # Adding Background to the main window
        l2 = QtWidgets.QLabel(w) # Adds Label directly to the window
        l2.setPixmap(QtGui.QPixmap("Space.jpg")) # imports the image to use
        l2.move(120, 90) # moves the position of the image in the window
        
        # Adding text to the main window
        l1 = QtWidgets.QLabel() # Adds Label directly to the window
        l1.setText("hello world") # Setting the text for the label
        #l1.move(130, 20) # Setting the position of the label
    
        # Adding Image to the main window
        l2 = QtWidgets.QLabel(w) # Adds Label directly to the window
        l2.setPixmap(QtGui.QPixmap("Space.jpg")) # imports the image to use
        l2.move(120, 90) # moves the position of the image in the window
    
        # adding a button to the window
        btn_1 = QtWidgets.QPushButton() # Adds button directly to the window
        btn_1.setText("push me") # Setting the text for the button
        #btn_1.move(120, 120) # Setting the position of the button
    
        # Adding a horizontal layout
        h_box = QtWidgets.QHBoxLayout()
        h_box.addStretch()
        h_box.addWidget(l1)
        h_box.addStretch()
    
        # Adding a Verticel layout
        v_box = QtWidgets.QVBoxLayout()
        v_box.addWidget(btn_1)
        v_box.addLayout(h_box)
    
        w.setLayout(v_box)
        w.setWindowTitle('AI Tester') # Setting the title of the main window
        w.setGeometry(100, 100, 300, 200) # Setting the size of the Window with the first 2 ints and the position of the window on the screen with the last 2 ints
        """
        # Setting fixed window Size
        self.setFixedSize(self.width, self.height)
        
        # Setting initial rates
        self.learn_value = 25
        self.poison_value = 30
        self.rich_value = 60
        
        # Initializing all Labels
        self.learn_label = QLabel("Learning Rate") # Adds Label directly to the window
        self.poison_label = QLabel("Poison Drop Rate") # Adds Label directly to the window
        self.rich_label = QLabel("Rich Resource Drop Rate") # Adds Label directly to the window
        
        # Initializing all the Sliders
        self.learn_slide = QSlider(Qt.Horizontal)
        self.learn_slide.setMinimum(0)
        self.learn_slide.setMaximum(30)
        self.learn_slide.setValue(self.learn_value)
        self.learn_slide.setTickInterval(10)
        self.learn_slide.setTickPosition(QSlider.TicksBelow)
        
        self.poison_slide = QSlider(Qt.Horizontal)
        self.poison_slide.setMinimum(0)
        self.poison_slide.setMaximum(100)
        self.poison_slide.setValue(self.poison_value)
        self.poison_slide.setTickInterval(10)
        self.poison_slide.setTickPosition(QSlider.TicksBelow)

        self.rich_slide = QSlider(Qt.Horizontal)
        self.rich_slide.setMinimum(0)
        self.rich_slide.setMaximum(100)
        self.rich_slide.setValue(self.rich_value)
        self.rich_slide.setTickInterval(10)
        self.rich_slide.setTickPosition(QSlider.TicksBelow)
        
        # Initializing all Line Editors
        self.learn_le = QLineEdit(str(self.learn_slide.value()))
        self.learn_le.setReadOnly(True)
        self.poison_le = QLineEdit(str(self.poison_slide.value()))
        self.poison_le.setReadOnly(True)
        self.rich_le = QLineEdit(str(self.rich_slide.value()))
        self.rich_le.setReadOnly(True)
        
        # Initializing all Buttons
        self.update_btn = QPushButton('Update')
        self.reset_btn = QPushButton('Reset')
        
        # Creating the Vertical box for the sliders
        v_box = QVBoxLayout()
        v_box.addWidget(self.learn_label)
        v_box.addWidget(self.learn_slide)
        v_box.addWidget(self.learn_le)
        
        v_box.addWidget(self.poison_label)
        v_box.addWidget(self.poison_slide)
        v_box.addWidget(self.poison_le)
        
        v_box.addWidget(self.rich_label)
        v_box.addWidget(self.rich_slide)
        v_box.addWidget(self.rich_le)
        
        v_box.addWidget(self.update_btn)
        v_box.addWidget(self.reset_btn)

        # Adding Background to the main window
        background = QLabel() # Adds Label directly to the window
        background.setPixmap(QPixmap("Space.jpg")) # imports the image to use
        background.move(0, 0) # moves the position of the image in the window
        
        
        h_box = QHBoxLayout()
        h_box.addWidget(background)
        h_box.addLayout(v_box)
        
        self.setLayout(h_box)
        self.setWindowTitle('PyQt5 Lesson 8')

        self.update_btn.clicked.connect(lambda: self.btn_clk(self.update_btn, 'Hello from Clear'))
        self.reset_btn.clicked.connect(lambda: self.btn_clk(self.reset_btn, 'Hello from Print'))
        self.learn_slide.valueChanged.connect(self.learn_change)
        self.poison_slide.valueChanged.connect(self.poison_change)
        self.rich_slide.valueChanged.connect(self.rich_change)

        
        self.show()
    
    def btn_clk(self, b, string):
        if b.text() == 'Print':
            print(self.le.text())
        else:
            self.le.clear()
        print(string)

    def learn_change(self):
        self.learn_value = str(self.learn_slide.value())
        self.learn_le.setText(self.learn_value)
        
    def poison_change(self):
        self.poison_value = str(self.poison_slide.value())
        self.poison_le.setText(self.poison_value)
        
    def rich_change(self):
        self.rich_value = str(self.rich_slide.value())
        self.rich_le.setText(self.rich_value)
    

app = QApplication(sys.argv)
a_window = MainScreen()
sys.exit(app.exec_())
```
